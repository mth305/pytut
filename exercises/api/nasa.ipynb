```python
asteroids = d["near_earth_objects"]

diameters = []

for a in asteroids:
    dia = a["estimated_diameter"]["kilometers"]["estimated_diameter_max"]
    diameters.append(dia)

plt.plot(diameters)
plt.show()
```
